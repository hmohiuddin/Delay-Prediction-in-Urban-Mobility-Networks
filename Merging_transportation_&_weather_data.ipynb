# Merging Transportation data and Weather data

```python
import pandas as pd
```

```python
df_w = pd.read_csv('Weather.csv')
```

```python
df_w.head()
```

```python
df_w = df_w.drop('Date_time',1)
```

```python
df_w['Datetime'] = pd.to_datetime(df_w['date'].apply(str)+' '+df_w['time'])
```

```python
df_w.head()
```

```python
df_w.to_csv('Weather_final.csv', index=False)
```

```python
df_bus = pd.read_csv('/content/BusData_Delay_full.csv') 
```

```python
df_bus.columns
```

```python
df_bus['Datetime'] = pd.to_datetime(df_bus['DT_actdpt'])
```

```python
df_bus_c = df_bus.sort_values('Datetime')
```

```python
df_w_c = df_w.sort_values('Datetime')
```

```python
df = pd.merge_asof(df_bus_c, df_w_c, on='Datetime', direction= 'nearest')
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.to_csv('df_final.csv', index=False)
```
